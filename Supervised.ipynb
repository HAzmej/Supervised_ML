```python
label=label.astype(int)
label
```

### 2.3.2 Spliiting the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
features_model = pd.DataFrame(features_scaled, columns=features.columns)
```

```python
X_train, X_test, y_train ,y_test = train_test_split(features_model, label, test_size=0.2, random_state=42)
```

# 3. Creating a model

## 3.1 Choix du model

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import GradientBoostingClassifier , RandomForestClassifier, AdaBoostClassifier
```

```python
models= { 
    'GradientBoosting': GradientBoostingClassifier(),
    'RandomForest': RandomForestClassifier(),
    'AdaBoost': AdaBoostClassifier()
        } 
    
```

## 3.2 Hyperparametres

```python
param_grid={
    'GradientBoosting':{
        'n_estimators':[100,200],
        'loss':['log_loss','exponential'],
        'learning_rate':[0.1,0.2,0.5],
        'max_depth':[1,2,5]
    },
    'RandomForest':{
        'n_estimators':[50,100,200],
        'criterion':['log_loss','gini'],
        'max_depth':[1,2,5,10],
        'min_samples_split':[2,5,10],
        'max_features':['sqrt','log2']
    },
    'AdaBoost':{
        'n_estimators':[50,100,150,200],
        'learning_rate':[0.1,0.2,0.5]
    }
        
}
```

```python
import time 

for m,s in models.items():
    a=time.time()
    model=GridSearchCV(s,param_grid[m])
    model.fit(X_train,y_train)
    b=time.time()
    y_predict=model.predict(X_test)
    print(f"Meilleurs paramètres : {model.best_params_}")
    print(f"Meilleur score (validation) : {model.best_score_:.4f}")
    print(f"Accuracy:{(accuracy_score(y_test, y_predict)*100):.2f}%\n")
    print("Classification Report:\n", classification_report(y_test, y_predict))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_predict))
    print("Temps d'execution:\n", b-a)
```
